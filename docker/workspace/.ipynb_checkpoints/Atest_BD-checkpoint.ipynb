```python
-- Создание таблицы в Hive/Hue
CREATE TABLE hue__tmp_ab_nyc_2019 (
    id INT,
    name STRING,
    host_id INT,
    host_name STRING,
    neighbourhood_group STRING,
    neighbourhood STRING,
    latitude FLOAT,
    longitude FLOAT,
    room_type STRING,
    price FLOAT,
    minimum_nights INT,
    number_of_reviews INT,
    last_review STRING,
    reviews_per_month FLOAT,
    calculated_host_listings_count INT,
    availability_365 INT
) ROW FORMAT DELIMITED FIELDS TERMINATED BY ',' STORED AS TEXTFILE;

-- Загрузка данных в таблицу
LOAD DATA LOCAL INPATH '/path/to/your/hue__tmp_ab_nyc_2019' INTO TABLE airbnb_data;

-- Вычисление среднего и дисперсии
SELECT AVG(price) AS average_price, VARIANCE(price) AS variance_price FROM airbnb_data;
```

```python
#!/usr/bin/env python mapper.py
import sys

# Чтение данных построчно
for line in sys.stdin:
    line = line.strip()
    fields = line.split(',')
    try:
        price = float(fields[9])  # Предполагаем, что price находится на 10-м месте
        print(f"{price}\t1")
    except ValueError:
        continue
```

```python
#!/usr/bin/env python reducer.py
import sys

total_sum = 0
total_count = 0
squared_sum = 0

for line in sys.stdin:
    line = line.strip()
    price, count = map(float, line.split('\t'))
    total_sum += price
    total_count += count
    squared_sum += price ** 2

# Расчет среднего и дисперсии
if total_count > 0:
    average_price = total_sum / total_count
    variance_price = (squared_sum / total_count) - (average_price ** 2)
    print(f"Average: {average_price}, Variance: {variance_price}")
```

##Запуск MapReduce
После того как вы создали mapper.py и reducer.py, вы можете запустить их на вашем кластере Hadoop с помощью команды:

```python
# bash
cat B_NYC_2019.csv python mapper.py sort -k1,1 python reducer.py
```
